### **Distribuci√≥n de evasi√≥n**

En este paso, el objetivo es comprender c√≥mo est√° distribuida la variable "churn" (evasi√≥n) entre los clientes. Utiliza gr√°ficos para visualizar la proporci√≥n de clientes que permanecieron y los que se dieron de baja.

```python
# Filtramos los datos, excluyendo a los clientes cuyo estado de permanencia se desconoce,
# ya que estos tienen un porcentaje bajo que no influye en el an√°lisis
filtroChurn = datos['Churn'] != -1 # ['No' 'Yes' 'Desconocido']
datos_churn_filtrados = datos[filtroChurn]
print(f'\nCantidad de clientes:\n{len(datos_churn_filtrados)}')
```

```python
datos_churn_filtrados.describe()
```

```python
print(f"Cantidad de clientes que permanecen: {datos_churn_filtrados[datos_churn_filtrados['Churn'] == 0].shape[0]}")
print(f"Cantidad de clientes que se fueron: {datos_churn_filtrados[datos_churn_filtrados['Churn'] == 1].shape[0]}")
```

```python
# Graficando barras de clientes por permanencia

import matplotlib.pyplot as plt

# Crear el gr√°fico y guardar el eje
ax = datos_churn_filtrados['Churn'].map({0: 'Permanecen', 1: 'Evadieron'}).value_counts().plot(
    kind='bar',
    color=['lightgreen', 'lightcoral']
)

# T√≠tulo y etiquetas
plt.title('Distribuci√≥n por permanencia', fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.xlabel('Clientes que permanecen vs que evadieron', fontdict={'fontsize': 12})
plt.ylabel('Cantidad', fontdict={'fontsize': 12})

plt.xticks(ticks=[0, 1], labels=['Permanecen', 'Evadieron'], rotation=0)

# Agregar los valores encima de cada barra
for i, p in enumerate(ax.patches):
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2,
            height + 10,              # un poco arriba de la barra
            int(height),              # el n√∫mero
            ha='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

### **Recuento de evasi√≥n por variables categ√≥ricas**

Ahora, exploraremos c√≥mo se distribuye la **evasi√≥n** seg√∫n variables categ√≥ricas, **como g√©nero, tipo de contrato, m√©todo de pago, entre otras**.

Este an√°lisis puede revelar **patrones interesantes**, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudar√° a **orientar acciones estrat√©gicas**.

```python
# Evasi√≥n seg√∫n g√©nero
# Crear una tabla pivote: filas = gender, columnas = Churn
churn_genero = datos_churn_filtrados.pivot_table(
    index='gender',
    columns='Churn',
    aggfunc='size',
    fill_value=0
)

# Mapear valores si est√°n codificados
churn_genero.index = churn_genero.index.map({0: 'Feminino', 1: 'Masculino'})
churn_genero.columns = ['Permanecen', 'Evadieron']

# Gr√°fico de barras agrupadas
ax = churn_genero.plot(kind='bar', color=['lightgreen', 'lightcoral'])

plt.title('Distribuci√≥n por g√©nero y permanencia', fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.xlabel('G√©nero', fontdict={'fontsize': 12})
plt.ylabel('Cantidad', fontdict={'fontsize': 12})
ax.legend(title='', loc='upper right')#, bbox_to_anchor=(1.18, 1.02))

plt.xticks(rotation=0)

# Mostrar valores encima
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Evasi√≥n seg√∫n tipo de contrato
# Crear una tabla pivote: filas = Contract, columnas = Churn
churn_contrato = datos_churn_filtrados.pivot_table(
    index='Contract',
    columns='Churn',
    aggfunc='size',
    fill_value=0
)

# Mapear valores si est√°n codificados
churn_contrato.index = churn_contrato.index.map({0: 'Mes a mes', 1: 'Por un a√±o', 2: 'Por dos a√±os'})
churn_contrato.columns = ['Permanecen', 'Evadieron']

# Gr√°fico de barras agrupadas
ax = churn_contrato.plot(kind='bar', color=['lightgreen', 'lightcoral'])

plt.title('Distribuci√≥n por tipo de contrato y permanencia', fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.xlabel('Tipo de contrato', fontdict={'fontsize': 12})
plt.ylabel('Cantidad', fontdict={'fontsize': 12})
ax.legend(title='', loc='upper right')

plt.xticks(rotation=0)

# Mostrar valores encima
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Evasi√≥n seg√∫n m√©todo de pago
import matplotlib.pyplot as plt
import seaborn as sns

# Crear una tabla pivote: filas = Contract, columnas = Churn
churn_metodo = datos_churn_filtrados.pivot_table(
    index='PaymentMethod',
    columns='Churn',
    aggfunc='size',
    fill_value=0
)

# Mapear valores si est√°n codificados
churn_metodo.index = churn_metodo.index.map({0: 'Electronic check', 1: 'Mailed check', 2:'Bank transfer (automatic)', 3: 'Credit card (automatic)'})
churn_metodo.columns = ['Permanecen', 'Evadieron']

# Gr√°fico de barras agrupadas
ax = churn_metodo.plot(kind='bar', color=['lightgreen', 'lightcoral'])

plt.title('Distribuci√≥n por m√©todo de pago y permanencia', fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.xlabel('M√©todo de pago', fontdict={'fontsize': 12})
plt.ylabel('Cantidad', fontdict={'fontsize': 12})
ax.legend(title='', loc='upper right')

plt.xticks(rotation=30)

# Mostrar valores encima
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10)

plt.tight_layout()
plt.show()

```

```python
# Tiempo de contrato
import pandas as pd
import plotly.express as px

# Agrupar la permanencia (tenure) en intervalos de 6 meses
datos_churn_filtrados_interv = datos_churn_filtrados.copy()
datos_churn_filtrados_interv['tenure_interval'] = pd.cut(datos_churn_filtrados_interv['tenure'], bins=range(0, 80, 6))

# Calcular la cantidad de personas que evadieron (Churn == 1)
grupo = datos_churn_filtrados_interv[datos_churn_filtrados_interv['Churn'] == 1] \
    .groupby(['tenure_interval', 'SeniorCitizen'], observed=True) \
    .size().reset_index(name='ChurnCount')

# Convertir intervalos a string para eje X
grupo['tenure_interval'] = grupo['tenure_interval'].astype(str)

# Mapear valores binarios a etiquetas
grupo['AdultoMayor'] = grupo['SeniorCitizen'].map({0: 'No', 1: 'S√≠'})

# Crear gr√°fico interactivo
fig = px.line(
    grupo,
    x='tenure_interval',
    y='ChurnCount',
    color='AdultoMayor',
    markers=True,
    labels={
        'tenure_interval': 'Meses de contrato (agrupados)',
        'ChurnCount': 'Cantidad de evasiones'
    },
    title='Cantidad de evasiones seg√∫n permanencia y edad (adulto mayor)',
    color_discrete_map={
        'No': 'skyblue',
        'S√≠': 'sandybrown'
    },
    hover_data={'AdultoMayor': False}  # Oculta esta columna del tooltip
)

# Ajustes visuales
fig.update_layout(
    xaxis_tickangle=-45,
    legend_title='Adulto mayor:',
    legend=dict(yanchor="top", y=0.99, xanchor="right", x=0.99)
)

plt.show()

fig.write_html("evasiones_permanencia_edad.html")
from google.colab import files
files.download("evasiones_permanencia_edad.html")
```
